```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Metro_Interstate_Traffic_Volume.csv')
```

```python
df['Dates'] = pd.to_datetime(df['date_time']).dt.date
df['Time'] = pd.to_datetime(df['date_time']).dt.time
```

```python
df.head()
```

```python
df.drop_duplicates(subset = 'date_time' , keep = 'first',inplace = True)
```

```python
df.drop(['date_time'] , axis = 1 , inplace = True)
```

```python
df.drop(['weather_description'] , axis = 1 , inplace = True)
```

```python
import datetime
import calendar
df['Day'] = [calendar.day_name[datetime.datetime.strptime(str(df['Dates'][i]), '%Y-%m-%d').weekday()] for i in df.index]
df
```

```python
df = df[['holiday','Time','Day','traffic_volume' ]]
```
